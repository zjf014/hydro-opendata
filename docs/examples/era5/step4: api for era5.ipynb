# 读取era5-land数据

```python
from wis_s3api import era5
import numpy as np
```

```python
start_time = np.datetime64('2021-01-01T01:00:00.000000000')
end_time = np.datetime64('2021-01-31T00:00:00.000000000')
```

## 根据经纬度范围

```python
bbox = (121,38,122,40)
```

```python
ds = era5.open(['Total precipitation','10 metre U wind component'],start_time=start_time,end_time=end_time,bbox=bbox)
ds
```

## 根据矢量数据范围

```python
shp = 'test.shp'
```

```python
ds = era5.from_shp(['Total precipitation','10 metre U wind component'],start_time=start_time,end_time=end_time,shp=shp)
ds
```

### 生成nc文件

#### hourly

```python
era5.to_netcdf(['Total precipitation','10 metre U wind component'],start_time=start_time,end_time=end_time,shp='test.shp',save_file='hourly.nc')
```

#### daily

```python
era5.to_netcdf(['Total precipitation','10 metre U wind component'],start_time=start_time,end_time=end_time,shp='test.shp',resolution='daily',save_file='daily.nc')
```

#### 6-hourly

```python
era5.to_netcdf(['Total precipitation','10 metre U wind component'],start_time=start_time,end_time=end_time,shp='test.shp',resolution='6-hourly',save_file='6-hourly.nc')
```
